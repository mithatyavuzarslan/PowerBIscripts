```python
pd.options.mode.chained_assignment = None 

df=pd.read_json("https://data.seattle.gov/resource/tazs-3rd5.json?$limit=100000&$where=report_datetime%3E%222024-01-01T00:00:00.000%22&$$app_token=OgpNzb9AWOYtUsK8x5pvXHepb")
```

```python
df['report_datetime']=pd.to_datetime(df['report_datetime'])
df['offense_start_datetime']=pd.to_datetime(df['offense_start_datetime'])
```

```python
df=df[(df['latitude']!=0.000000)| (df['latitude']!=0.0)]
```

```python
df.loc[:, 'month'] = pd.Series(df['report_datetime'].dt.month)
df.loc[:, 'year'] = pd.Series(df['report_datetime'].dt.year)
df.loc[:, 'day'] = pd.Series(df['report_datetime'].dt.day)
```

```python
df.sort_values(by='report_datetime',inplace=True)
```

```python
def format_decimal(value):
    return str(value).replace('.', ',')
```

```python
long_formatted=df['longitude'].apply(format_decimal)
lat_formatted=df['latitude'].apply(format_decimal)

df['longitude']=pd.Series(long_formatted,dtype="float64")
df['latitude']=pd.Series(lat_formatted,dtype="float64")

print(df)
```
